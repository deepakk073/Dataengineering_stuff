```
import apache_beam as beam
def retTuple(element):
  tup=element.split(',')
  return (tup[0], tup[1:])

p = beam.Pipeline()
dep_rows = (p|"Reading Department Files">> beam.io.ReadFromText('dept_data.txt')
            |"pair _employee">>beam.Map(retTuple)
          #  |"print">>beam.Map(print)

)

loc_rows = (
          p|"read location files">>beam.io.ReadFromText("location.txt")
          |beam.Map(retTuple)
        #  |beam.Map(print)
)

result =({"dep_data":dep_rows,'loc_data':loc_rows}
         |beam.CoGroupByKey()
         |beam.io.WriteToText('result')
)
p.run()
!{('head -n 22 result-*')}
```
